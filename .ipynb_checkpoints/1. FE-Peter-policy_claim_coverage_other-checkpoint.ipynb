# 下面炸裂注意

```python
df_policy_coverage=pd.DataFrame()
df_policy_number_unique = df_policy['Policy_Number'].unique()
all_number = len(df_policy_number_unique)
print(all_number)

appended_df_policy_coverage_each=[]
j=1
t0 = time.time()
for policy_number in df_policy_number_unique:
    if j % 1000 == 0:
        t1 = time.time()
        print("Processing file {} ({}%)".format(j, 100*j//all_number), end="")
            
    df_temp = df_policy.loc[df_policy['Policy_Number'] == policy_number]
    df_policy_coverage_each=pd.DataFrame()
    total_Premium = 0
    for i in df_temp.index:
        #print(type(df_temp.loc[i,'Policy_Number']))
        df_policy_coverage_each.at[0,'Policy_Number'] = df_temp.at[i,'Policy_Number']
        #print(df_policy_coverage_each['Policy_Number'])
        df_policy_coverage_each.at[0,df_temp.at[i,'Insurance_Coverage']+'_Insured_Amount1'] = df_temp.at[i,'Insured_Amount1']
        df_policy_coverage_each.at[0,df_temp.at[i,'Insurance_Coverage']+'_Insured_Amount2'] = df_temp.at[i,'Insured_Amount2']
        df_policy_coverage_each.at[0,df_temp.at[i,'Insurance_Coverage']+'_Insured_Amount3'] = df_temp.at[i,'Insured_Amount3']
        df_policy_coverage_each.at[0,df_temp.at[i,'Insurance_Coverage']+'_Premium'] = df_temp.at[i,'Premium']
        total_Premium = total_Premium + df_temp.at[i,'Premium']
        df_policy_coverage_each['total_Premium'] = total_Premium
        #print(df_policy_coverage_each)
    appended_df_policy_coverage_each.append(df_policy_coverage_each)
            
    if j % 1000 == 0:
        t2 = time.time()
        print(" {}s (total: {}s)".format((t2-t1)*1000, t2-t0))
    j+=1
    
df_policy_coverage = pd.concat(appended_df_policy_coverage_each)
```

```python
df_policy_coverage.shape
```

```python
cols = df_policy_coverage.columns.tolist()
cols = cols[-2:] + cols[:-2]

df_policy_coverage = df_policy_coverage[cols]

df_policy_coverage.fillna(0, inplace=True)
df_policy_coverage = df_policy_coverage.reset_index(drop=True)

df_policy_coverage.head(10)
```
