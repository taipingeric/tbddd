```python
import pandas as pd
import numpy as np
df_policy_raw = pd.read_csv('../Tbrain_Insurance/policy_claim/policy_0702.csv')
df_claim_raw = pd.read_csv('../Tbrain_Insurance/policy_claim/claim_0702.csv')
df_training_raw = pd.read_csv('../Tbrain_Insurance/training-set.csv')
df_testing_raw = pd.read_csv('../Tbrain_Insurance/testing-set.csv')
```

```python
df_policy_claim = df_policy_raw.merge(df_claim_raw, left_on = ['Policy_Number', 'Insurance_Coverage'], right_on = ['Policy_Number', 'Coverage'], how= 'left')
df_policy_claim = df_policy_claim.merge(df_training_raw, on = ['Policy_Number'], how= 'left')
df_policy_claim = df_policy_claim.merge(df_testing_raw, on = ['Policy_Number'], how= 'left')
```

```python
df_policy_claim['Next_Premium_x'] = df_policy_claim['Next_Premium_x'].fillna(0)
df_policy_claim['Next_Premium_y'] = df_policy_claim['Next_Premium_y'].fillna(0)
df_policy_claim['Next_Premium'] = df_policy_claim['Next_Premium_x'] + df_policy_claim['Next_Premium_y']
df_policy_claim['Next_Premium'].head()
```

```python
df_policy_claim_grouped = df_policy_claim.groupby(['Policy_Number']).first().reset_index()
```
